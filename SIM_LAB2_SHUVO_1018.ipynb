1st question:

```python
import matplotlib.pyplot as plt

x = np.random.rand(15)
y = np.random.rand(15)

plt.plot(range(15), x, marker='o', label='Vector X')
plt.plot(range(15), y, marker='s', label='Vector Y')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Comparison of Two Random Vectors')
plt.legend()
plt.grid(True)
plt.show()
```

2nd Question:

```python
import seaborn as sns

matrix_4x4 = np.random.rand(4, 4)

sns.heatmap(matrix_4x4, annot=True, cmap='viridis', linewidths=0.5, linecolor='black')
plt.title("Heatmap of 4x4 Random Matrix")
plt.xlabel("Columns")
plt.ylabel("Rows")
plt.show()
```

3rd question:

```python
A = np.random.randint(1, 10, (4, 4))
B = np.random.randint(1, 10, (4, 4))

sum_matrix = A + B
diff_matrix = A - B
prod_matrix = A @ B

# Flatten matrices for plotting
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot Addition
axes[0].bar(range(16), sum_matrix.flatten(), color='blue')
axes[0].set_title("Matrix Addition")
axes[0].set_xlabel("Element Index")
axes[0].set_ylabel("Value")

# Plot Subtraction
axes[1].bar(range(16), diff_matrix.flatten(), color='green')
axes[1].set_title("Matrix Subtraction")
axes[1].set_xlabel("Element Index")
axes[1].set_ylabel("Value")

# Plot Multiplication
axes[2].bar(range(16), prod_matrix.flatten(), color='red')
axes[2].set_title("Matrix Multiplication")
axes[2].set_xlabel("Element Index")
axes[2].set_ylabel("Value")

plt.tight_layout()
plt.show()
```
