```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt


data = np.array([5.2, 5.4, 5.1, 5.6, 5.3, 5.8])
population_mean = 5.5


t_statistic, p_value = stats.ttest_1samp(data, population_mean)


print(f'Sample Mean: {np.mean(data):.3f}')
print(f'Hypothesized Population Mean: {population_mean:.3f}')
print(f't-Statistic: {t_statistic:.3f}')
print(f'Degrees of Freedom: {len(data) - 1}')
print(f'p-Value: {p_value:.4f}')


alpha = 0.05
if p_value < alpha:
    print(f'Reject the null hypothesis. The average is significantly different from {population_mean:.3f}.')
else:
    print(f'Fail to reject the null hypothesis. The average is not significantly different from {population_mean:.3f}.')


plt.figure()
plt.bar(1, np.mean(data), color="skyblue", label='Sample Mean')
plt.errorbar(1, np.mean(data), yerr=np.std(data, ddof=1), fmt='k', linewidth=1.5)  # use sample std dev
plt.axhline(y=population_mean, color='r', linestyle='--', linewidth=1.5, label='Hypothesized Mean')
plt.title('Sample Mean vs Hypothesized Mean')
plt.xlim(0.5, 1.5)
plt.xlabel('Sample')
plt.ylabel('Mean Value')
plt.legend()
plt.grid(True)
plt.show()
```
